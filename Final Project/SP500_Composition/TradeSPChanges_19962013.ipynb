# Trading SP500 composition changes
We wanted to see if one could take advantage of the expected rise(fall) in the stock share price from announcement of the composition change to the actual effective change date.

It is important between types of index changes when conducting research such as this one. The most important events are when the index committee decides to make a change based on their selection criteria. For example, when a company outside of the index has grown rapidly and passed a company in the index in market cap. In those cases passive funds will need to add the company entering the index to their portfolio and get rid of the company leaving the index. This creates trading opportunities on both sides. 

Other index changes will usually only require passive funds to trade either the entering or leaving company and thus only create a supply and demand imbalance for one company. For example, when a company that is a part of the index is taken over by another or taken of the market. Another example is when a company that is already a part of the index spins off another company such that both parts are still large enough to be a part of the index.

# Strategies
We created a framework that allows us to test dispersion trading strategies on the index change events. A fixed amount is allocated to each trade with options for offsetting the open and close of the trade relative to the announcement and change dates. 

## Strategy 0


Trade all stocks entering the S&P 500 index

Open trade on the morning after the announcement is made. Short the index through the SPY ETF and use proceeds to go long the stock entering the index.

Close trade at close of trading on the effective change date. Buy SPY back and sell the stock

Results: Very profitable from 1996 until the early 2000's after that the returns slow down and since the crisis this hasn't been profitable at all.

## Strategy 1
Pair trade stocks entering and leaving the index when possible, i.e. when unforced changes are made by index committee.

Open trade on the morning after the announcement is made. Short the stock leaving the index and use the proceeds to buy the stock entering the index.

Close trade at close of trading on the effective change date. Buy short stock back and sell the long stock

Results: Similar history of profitability. More profitable on average than the previous strategy since it is actually more profitable to trade stock leaving the index than those entering the index. This is not that surprising since the stocks leaving index are usally leaving because their stock price has been declining. 

```python
strategy = 1
fixed_trade_amount = 10000
timing = False
close_offset=0
t_cost = 0.0010
file_name = 'TradeLogs/Strategy1_19962013.txt'
results = run_trading_strategy(data, strategy=strategy, timing=timing, t_cost=t_cost, log_file=file_name, close_offset=close_offset)
```
