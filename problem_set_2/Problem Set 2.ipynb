```python
data = nbprint_and_query('q4.sql')
data
```

```python
import pandas as pd

groups = data.groupby(['Date', 'CallPut', 'Expiration'])

for name, group in groups:
    monotone_function  = lambda x: pd.algos.is_monotonic_float64(x, False)
    is_inc, is_dec = monotone_function(group.sort_values(by='Strike')['MBBO'].values)
    call_flip = name[1] == 'C' and not is_dec
    put_flip = name[1] == 'P' and not is_inc
    if call_flip or put_flip:
        print("Price Flip on {} for Call/Put {}, expiring {}".format(
                name[0], name[1], name[2]
            ))
```
