```python
import sys
sys.path.append('../../utils/')

from query import query_dataframe
from trading_days import TradingDay

date_format = '%Y-%m-%d'

equity_query = '''
SELECT sp.ClosePrice, sp.Date
FROM XFDATA.dbo.SECURITY_PRICE sp
INNER JOIN XFDATA.dbo.SECURITY_NAME sn on sp.SecurityID = sn.SecurityID
WHERE sn.Ticker = 'INTC'
  AND sp.Date BETWEEN '2011-01-10' AND '2011-02-14'
'''

options_query = '''
SELECT
  op.Date,
  op.OpenInterest,
  op.CallPut
FROM XFDATA.dbo.OPTION_PRICE_VIEW op
  INNER JOIN XFDATA.dbo.SECURITY_NAME sn ON op.SecurityID = sn.SecurityID

WHERE sn.Ticker = 'INTC'
      AND XF.dbo.formatStrike(op.Strike) = 21
      AND op.Date BETWEEN '2011-01-10' AND '2011-02-14'
      AND op.Expiration = '2011-02-19'
GROUP BY op.Date, op.OpenInterest, op.CallPut
'''
```

```python
equity_data = query_dataframe(equity_query)
options_data = query_dataframe(options_query)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline

f, ax1 = plt.subplots()
equity_data.plot(x='Date', y='ClosePrice', ax=ax1, label='Stock Price')

earnings = datetime(2011, 1, 14)
ax1.axvline(earnings - TradingDay(1), ls='dashed', color='g')
ax1.axvline(earnings, ls='dashed')
ax1.axvline(earnings + TradingDay(10), ls='dashed', color='r')
ax1.legend(loc=2)

ax2 = ax1.twinx()
options_data[options_data['CallPut'] == 'C'].plot(x='Date', y='OpenInterest', ax=ax2, label='Call Open Interest', color='g')
options_data[options_data['CallPut'] == 'P'].plot(x='Date', y='OpenInterest', ax=ax2, label='Put Open Interest', color='r')
ax2.set_ylim([10000, 120000])
ax2.legend(loc=4)

f.set_size_inches(12, 6)
ax1.set_title('INTC Earnings 1/14/2011')
```
